### FL for the new evaluation metric

```python
import matplotlib.pyplot as plt

def getX(l):
    return [p[0] for p in l]

def getY(l):
    return [p[1] for p in l]

c1 = [[-0.25,1.25], [0,1], [0.25,1.25]]
c2 = [[-1.25,-0.25], [-1,0], [-1.25,0.25]]
c3 = [[1.25, -0.25], [1,0], [1.25, 0.25]]

q1 = [c1[1], c2[1], c3[1]]
q2 = [c1[0], c1[-1], c2[0], c2[-1], c3[0], c3[-1]]

plt.scatter(getX(q1), getY(q1), label='query 1')
plt.scatter(getX(q2), getY(q2), label='query 2')
plt.legend()
plt.show()
```

```python
from sklearn_extra.cluster import KMedoids
import numpy as np

X = np.asarray([*c1, *c2, *c3])
kmedoids = KMedoids(n_clusters=3, method='pam').fit(X)

plt.scatter(getX(kmedoids.cluster_centers_), getY(kmedoids.cluster_centers_))
plt.show()
```
