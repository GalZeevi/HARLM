### Database Papers From Selected Universities

```python
sql = '''
SELECT title,
       YEAR,
       o.name AS organization
FROM organization o,
     publication p,
     publication_organization po,
     domain_publication dp,
     DOMAIN d
WHERE p.pid = po.pid
  AND o.oid = po.oid
  AND dp.pid = p.pid
  AND dp.did = d.did
  AND d.name = 'Databases'
  AND o.name IN ('Tel Aviv University', 'University of Maryland', 'University of Michigan');
'''
asqp_rl.query_asqp(sql)
demonstrate_asqp_rl(asqp_rl)
```

### Papers From Selected Authors

```python
sql = '''
SELECT a.name,
       a.paper_count,
       a.citation_count,
       o.name AS "organization"
FROM author a,
     writes w,
     publication p,
     organization o,
     publication_organization po
WHERE a.aid = w.aid
  AND w.pid = p.pid
  AND p.pid = po.pid
  AND o.oid = po.oid
  AND a.paper_count < 100
  AND a.citation_count > 1000
  AND o.name = 'University of California San Diego'
  AND p.year > 2013;
'''
asqp_rl.query_asqp(sql)
demonstrate_asqp_rl(asqp_rl)
```
